# Calcular as distância entre as escolas

```python
df1 = dfx[dfx['ALUNOS_ID_ESCOLA'] != 31118672]
```

```python
dfx = df1[df1['ALUNOS_ID_ESCOLA'] != 31081485] 
```

```python
df2 = dfx.drop(columns = ['ALUNOS_ID_ESCOLA', 'MEDIA_FINAL'])
```

```python
scaler = StandardScaler()
standard_df2 = scaler.fit_transform(df2)
standard_df2
```

```python
distance_matrix = distance.cdist(standard_df2, standard_df2, 'euclidean')
distance_matrix
```

```python
matrix_distance_sort = np.argsort(distance_matrix, axis=1)
matrix_distance_sort
```

```python
upper_triangle_dist = distance_matrix[np.triu_indices_from(distance_matrix, k=1)]

# Calcula a diferença absoluta das notas para cada par
media_final = dfx['MEDIA_FINAL'].values
diffs_notas = []
for i in range(len(media_final)):
    for j in range(i + 1, len(media_final)):
        diffs_notas.append(abs(media_final[i] - media_final[j]))
diffs_notas = np.array(diffs_notas)
```

```python
# Limites para as distâncias das características
dist_q1 = np.quantile(upper_triangle_dist, 0.07) # Limite para "semelhantes"
dist_q3 = np.quantile(upper_triangle_dist, 0.99) # Limite para "diferentes"

# Limites para as diferenças de notas
notas_q1 = np.quantile(diffs_notas, 0.07) # Limite para "semelhantes"
notas_q3 = np.quantile(diffs_notas, 0.99) # Limite para "diferentes"

print(f"Distância para ser 'semelhante' deve ser < {dist_q1:.2f}")
print(f"Distância para ser 'diferente' deve ser > {dist_q3:.2f}")
print(f"Diferença de nota para ser 'semelhante' deve ser < {notas_q1:.2f}")
print(f"Diferença de nota para ser 'diferente' deve ser > {notas_q3:.2f}")
```

```python
semelhantes_caracteristicas_notas_diferentes = []
diferentes_caracteristicas_notas_semelhantes = []

ids = dfx['ALUNOS_ID_ESCOLA'].values
media = dfx['MEDIA_FINAL'].values

for i in range(len(dfx)):
    for j in range(i + 1, len(dfx)):
        distancia_ij = distance_matrix[i, j]
        diff_nota_ij = abs(media[i] - media[j])

        # Grupo 1: Características semelhantes, notas diferentes
        if distancia_ij < dist_q1 and diff_nota_ij > notas_q3:
            semelhantes_caracteristicas_notas_diferentes.append((ids[i], ids[j], media[i], media[j], distancia_ij))

        # Grupo 2: Características diferentes, notas semelhantes
        if distancia_ij > dist_q3 and diff_nota_ij < notas_q1:
            diferentes_caracteristicas_notas_semelhantes.append((ids[i], ids[j], media[i], media[j], distancia_ij))

# Criar os DataFrames finais
df_grupo1 = pd.DataFrame(semelhantes_caracteristicas_notas_diferentes, columns=['ID_i', 'ID_j', 'MEDIA_i', 'MEDIA_j', 'Distancia'])
df_grupo2 = pd.DataFrame(diferentes_caracteristicas_notas_semelhantes, columns=['ID_i', 'ID_j', 'MEDIA_i', 'MEDIA_j', 'Distancia'])
```
